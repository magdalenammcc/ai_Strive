```python
# Counting words line by line.
from collections import OrderedDict

def display_contents_file4(file_name):
    fp = open(file_name)
    count = 0
    list_of_words=[] #list where I am storing all the words of the text
    
    while True:
        buffer = fp.readline()   
        list_of_words+=buffer.split()
        if buffer == "":
            break     
        
    fp.close()  # as suggested by Jon, finish reading the file at once
    
    
    set_of_words=set(list_of_words)  #with a set, I have the keys of the dictionary
    
    dictionary = dict.fromkeys(set_of_words, 0) # the keys in the dictionary come from the set
    
    
    for item in dictionary: 
        dictionary[item]= list_of_words.count(item)
    
    final_dictionary = sorted(dictionary)
        
    return final_dictionary


  
    
display_contents_file4("Shakespeare.txt")
```
